```python
import cv2
import numpy
from matplotlib import pyplot

video = cv2.VideoCapture('flag.webm')

flag = ""

while (video.isOpened()):
    ret, frame = video.read()
    if ret:
        frame = numpy.array(frame)
        frame = frame[0:32, 0:32*8]
        pyplot.imshow(frame, interpolation="nearest")
        pyplot.show()
    else:
        break
```
